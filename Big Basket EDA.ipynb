```python
# Necessary library required.

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Data Collection and Understanding.

df=pd.read_csv('D:\\kaggle_practice\\EDA Datasets\\Big basket dataset\\BigBasket Products.csv')
df
```

```python
# First view of Dataset.

df.head()
```

```python
# Column Names from the dataframe.

df.columns
```

```python
# Renaming the name of columns.

columns_mapping={'product':'Product','category':'Category','sub_category':'Sub_Category','brand':'Brand','sale_price':'Sale_Price','market_price':'Market_Price','type':'Type','rating':'Rating','description':'Description'}
df=df.rename(columns=columns_mapping)
```

```python
# Dataframe with updated column names.

df
```

```python
# Seperating the Numerical and Categorical columns in different list.

Numerical_columns=[]
Categorical_columns=[]
for column in df.columns:
    if df[column].dtype== 'object':
        Categorical_columns.append(column)
    else:
        Numerical_columns.append(column)
```

```python
# Numerical columns from the dataset.

Numerical_columns
```

```python
# Categorical Columns form the dataset.

Categorical_columns
```

```python
# Describing the Numerical columns of Dataset.

df.describe()
```

```python
# Checking the Information of Dataset.

df.info()
```

```python
# Checking the Null values from dataset.

df.isnull().sum()
```

```python
# Overall shape of dataset.

overall_data=df.shape[0]
overall_data
```

```python
# Dropping the Nan values from particular columns.

df_no_missing=df.dropna(subset=['Product', 'Brand', 'Description'], axis=0)
df_no_missing
```

```python
# Number of missing rows from dataset.

Missing_row=overall_data-df_no_missing.shape[0]
Missing_row
```

```python
# Percentage of missing rows from the dataset.

percentage_missing = (Missing_row / overall_data) *100
percentage_missing
```

```python
# After Removing Nan Values from the dataset.

df
```

```python
# Checking of Null Values after dropping the nan values from dataset.

df.isnull().sum()
```

```python
column_name = 'Rating' 

# Calculate the median for the specific column
median_value = df[column_name].median()
median_value
```

```python
# Fill missing values in the specific column with the median.

df[column_name].fillna(median_value, inplace=True)
```

```python
# Final check of missing values.

df.isnull().sum()
```

```python
# Removing duplicates rows from dataset.

clean_dataframe=df.drop_duplicates()
clean_dataframe
```

```python
# Distribution of sales price By using Histogram.

plt.figure(figsize=(8, 6))
plt.hist(df['Sale_Price'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Sale Prices')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Distribution of Products across categories.

plt.figure(figsize=(10, 6))
count = df['Category'].value_counts().plot(kind='bar', color='green', edgecolor='black')
count.set_xlabel('Category')
count.set_ylabel('Count')
count.set_title('Distribution of Products Across Categories')
plt.xticks(rotation=45, ha='right')
plt.show()

# Another way to find out the values by coding.
category_counts = df['Category'].value_counts()
most_represented_category = category_counts.idxmax()
least_represented_category = category_counts.idxmin()
```

```python
# Show top 10 brands of Big basket dataset.

top_brands = df['Brand'].value_counts().head(10).plot(kind='bar')
top_brands
```

```python
# Relationship between price and Category .

plt.figure(figsize=(8, 6))
plt.scatter(df['Sale_Price'], df['Category'], color='yellow', alpha=0.7)
plt.title('Relationship Between Price and Category')
plt.xlabel('Price')
plt.ylabel('Rating')
```

```python
# Relationship between price and Rating.

plt.figure(figsize=(8, 6))
plt.scatter(df['Sale_Price'], df['Rating'], color='violet', alpha=0.7)
plt.title('Relationship Between Price and Rating')
plt.xlabel('Price')
plt.ylabel('Rating')
```

```python
# Distribution of sales price By using Histogram.

plt.figure(figsize=(8, 6))
plt.hist(df['Rating'], bins=15, color='pink', edgecolor='black')
plt.title('Distribution of Sale Prices')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Box plot distribution of sale price with its categories.

plt.figure(figsize=(10, 6))
ax = sns.boxplot(x='Category', y='Sale_Price', data=df,)
ax.set_title('Comparison of Sale Prices Across Categories')
ax.set_xlabel('Category')
ax.set_ylabel('Sale Price')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Percentage of discount calculation for top 5 products.

df['Discount_Percentage'] = ((df['Market_Price'] - df['Sale_Price']) / df['Market_Price']) * 100
top_discounts = df.sort_values(by='Discount_Percentage', ascending=False).head(5)
top_discounts
```

```python
# category and sub category wise average rating.

average_ratings = df.groupby(['Category', 'Sub_Category'])['Rating'].mean().reset_index()
average_ratings
```

```python
# Top 10 brand by each category by rating

top_brands_by_category = df.groupby('Category')['Brand', 'Rating'].max().reset_index()
top_brands_by_category
```

```python
# Top brans by each category with its sale price.

top_brands_by_category = df.groupby('Category')['Brand', 'Sale_Price'].max().reset_index()
top_brands_by_category
```

```python
# Top 10 products price difference between its sale price and market price.

df['PriceDifference'] = df['Market_Price'] - df['Sale_Price']
sorted_df = df.sort_values(by='PriceDifference', ascending=False)
top_10_products = sorted_df.head(10)
top_10_products
```
